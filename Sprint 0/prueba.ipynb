```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.dates as mdates
import datetime
import random

# Generar fechas aleatorias para simular contribuciones
num_days = 365
start_date = datetime.date.today() - datetime.timedelta(days=num_days)
dates = [start_date + datetime.timedelta(days=i) for i in range(num_days)]
commits = [random.randint(0, 5) for _ in range(num_days)]  # Número aleatorio de commits por día

# Crear una figura
fig, ax = plt.subplots()

# Convertir fechas a números de días
dates_num = mdates.date2num(dates)

# Graficar cuadros para cada día
for i, (date, commit_count) in enumerate(zip(dates_num, commits)):
    color = mcolors.to_rgba('green', alpha=min(1, commit_count / 5))  # Color verde con transparencia según el número de commits
    # Calcula la coordenada x teniendo en cuenta que se muestren hacia la izquierda
    x = max(dates_num) - i
    ax.add_patch(plt.Rectangle((x, 0), 1, 1, color=color))

# Configuraciones adicionales
ax.set_xlim(min(dates_num), max(dates_num))
ax.set_ylim(0, 1)
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_minor_locator(mdates.DayLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
ax.yaxis.set_visible(False)
ax.set_title('Contribuciones diarias')

# Guardar la figura como imagen
plt.savefig('contribuciones.png')

# Mostrar la figura
plt.show()
```
